**If statement**

```python
# a=20
a=60
if a>50:
    print("This is if body")
print("This is outside the if block")
```

**If Else Statment**

```python
i=20
if i%2==0:
    print("I Is An Even Number")
else:
    print("I Is An Odd Number")
```

**Nested if Statement**

```python
c=50
if c<25:
    if c%2==0:
        print("Even number less than 25")
    else:
        print("Odd number less than 25")
else:
    print("Greater than 25")
```

**If Elif Else Statement**

```python
var='b'

if var=='a':
    print("this is the vowel a")
elif var=='e':
    print("this is the viwel e")
elif var=='i':
    print("this is the viwel i")
elif var=='o':
    print("this is the viwel o")
elif var=='u':
    print("this is the viwel u")
else:
    print("this is a consonant")
```
