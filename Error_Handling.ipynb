**Python Error Handling**

```python
# Error hadnling hepls your python program deal with unexepcted situations eithout crashing.
# it's how you gracefully manage problem like bad input,missing files ,or failed operations
```

1.Try-Except Blocks

```python
# Use try and except to catch and handles error
```

```python
# Syntax:
try:
    #code that might cause an error
except errorType:
    #code to run if that error occurs
```

```python
# Ex:

try:
    num=int(input("Enter a number :"))
    print("Your entered :" ,num)
except:
    print("Oops! That's not a valid number")
```

2. Catching Multiple Exception
   

```python
# You can handle different errors diffently
```

```python
# Ex :
try:
    num=int(input("Enter a number :"))
    result =10/num
    print("Result:",result)
    
except Exception as err :
    print(f"Sorry there is an err as {err}")
    
except ValueError:
    print("Please enter a valid number.")
```

3. The else and finally clauses

```python
# else: runs if no error occurs
# finally:always russ(used for cleanup)
```

```python
# Ex :
try:
    file = open("data.txt","r")
    contect=file.read()
    
except FileNotFoundError :
    print("File not Found")

else:
    print("File contect Loaded")

finally:
    print("Done trying to read the file")
```
