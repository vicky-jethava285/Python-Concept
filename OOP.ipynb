3. Instance Methods 

# Function inside classes that act on objects.

```python
class student :
    def __init__(self,name,age,cls):
        self.name = name
        self.age =age
        self.cls = cls
stud =student("vikas",21,'c1')
print(stud)
print(stud.cls)
print(stud.age)
```

4. Encapsulation

```python
# Encapsulation restrics access to certain parts of an object ,usually by using private variable or 
# methods(_var,__var).this keeps data safe and controlled.
```

```python
# Ex
class demo:
    def __int__(self):
        self.name = "vikas"   #public
        self._age =21   #protected
        self.__salary =50000   #private
    def show(self):
        print("inside the class")
        print("public :",self.name)
        print("protected :",self._age)
        print("private :",self.__salary)
dm =demo()    
print(dm)
```

5. Inheritance

```python
#Child class inherits features from parent class
```

```python
class Animal :
    def speak(self):
        print("Make a sound ")
class Dog(Animal):
    def bark(self):
        print("woof !")
d=Dog()
d.speak()  
d.bark()
```

6. Polymorphism


```python
# Same method name but differnet behovior
```

```python
class Bird:
    def sound(self):
        print("Chirp")
class Cat:
    def sound(self):
        print("Meow")
        
for animal in [Bird(),Cat()]:
    animal.sound()
```

Mini Project : Bank Account System in Python 

```python
class BankAccount:
    def __init__(self,name,balance = 0):
        self.name =name    #public
        self.__balance =balance  #private

    def deposit(self,amount):
        self.__balance += amount    #private   -  assigment operator (+=,-=)

    def withdrow(self,amount):
        if amount <= self.__balance:
            self.__balance -= amount    #private
        else:
            print("Insufficient Balance")

    def get_balance(self):
        return self.__balance

acc = BankAccount("vikas")
acc.deposit(100)
acc.withdrow(200)
print(acc.get_balance())
```
