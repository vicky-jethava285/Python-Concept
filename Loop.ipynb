1. While Loop

```python
# The while loop is used to repeat a section of code an unknown number of times until a specific condition is met  
```

```python
val=int(input("Enter a multiple of 7"))

while val%7 !=0:
    val=int(input("enter a multiple of 7"))
else:
    print("%d is a multiple of 7" %val)
```

2. For Loop

```python
# For loop is used to iterate over a sequence , which could be a list tuple array or a string
```

```python
x=[1,4,'vikas']
for i in x:
    print(i)
```

```python
x="vikasjethava"
for i in x:
    print(i)
```

3. Nested Loop

```python
# python allows loops to be nested in loops . nested loops could be:
#     for in for
#     while i while
#     for in while
#     while in for
```

```python
x=[[1,2,3],['a','b','c']]
for i in x:
    for j in i:
        print(j,end=" ")
    print()
```

5. Loop Control

```python
# Loop control statement alter the regular flow of a loop
```

```python
# break :transfers control to the statement right after the loop
```

```python
x="Hey. my name is vikas"
for i in x:
    if i == ".":
        break
    print(i,end=" ")
```

```python
# continue : skips the statements following it and return control to the begginning of the loops
```

```python
for i in[1,12,23,34]:
    if i==12:
        continue
    print(i)
```
