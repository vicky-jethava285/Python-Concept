1. Iterator & Generators 

```python
# Interators let you loop through data one item at a time .
# Generators are a memory - efficient way ato create iterators using yield
```

```python
# Interators Example

nums=["vikas","sanket","vishal"]
it = iter(nums)    #interators - iter
print(next(it))   #first 
print(next(it))   #second
print(next(it))   #third
```

```python
# Generators Example

def countdown(n):
    while n > 0:
        yield n
        n-=1
for n in countdown(10):
    print(n)
```

Challenge : Fibonacci Generator 

```python
def fibonacci():
    a,b = 0,1
    for _ in range(5):
        yield a
        a,b = b,a+b
for num in fibonacci():
    print(num)
```

2. Decorators 

```python
# Decorators wrap function to add extra behavior --like logging or timing --without changing the original code
```

**Step-by-step explanation**

def decorator(func):

You define a function named decorator that accepts one argument (func). This function is the decorator factory (a function that returns a wrapper for another function).

def wrap(): (inside decorator)

Inside decorator you define an inner function wrap. This will be the wrapper that runs additional code around the original function.

print("Before") inside wrap

When wrap runs it first prints "Before".

func() inside wrap

func refers to the original function passed into decorator (more on that in a second). This line calls the original function.

print("After") inside wrap

After the original function returns, the wrapper prints "After".

return wrap from decorator

The decorator returns the wrapper function object wrap. Important: it returns the function itself (not the result of calling it).

@decorator above def vikas():

This is syntactic sugar. Python does:

def vikas():
    print("Hello")
vikas = decorator(vikas)


So the name vikas is rebound to the wrap function returned by decorator. The original vikas function is now referenced inside that wrapper as func.

vikas() call

Because vikas now references wrap, calling vikas() runs wrap() which prints "Before", calls the original vikas() (which prints "Hello"), then prints "After".

```python
# Example :
def decorator(func):
    def wrap():
        print("Before")
        func()
        print("After")
    return wrap

@decorator
def vikas():
    print("Hello")
vikas()
```

Challenge : Time Decorator 

```python
import time
def timer(func):
    def wrap():
        start = time.time()
        func()
        end = time.time()
        print(f"Time :{end - start :.2f}s")
    return wrap
    
@timer
def slow_func():
    time.sleep(3)
    print("Done !")3
slow_func()
```

3. Context Managers

```python
# Use with to manage resources like files or database connections safely and cleanly 
```

```python
# Example :
with open("data.txt","r")as file:
    content = file.read()
    print(content)
```

Custom Context Manager

```python
class OpenFile:
    def __init__(self,filename):
        self.filename =filename
        
    def __enter__(self):
        self.file=open(self.filename,"r")
        return self.file
        
    def __exit__(self,exc_type,exc_val,exc_tb):
        self.file.close()

with OpenFile("data.txt")as f:
    print(f.read())
```

4. Regular Expression (Regex)

```python
# Regex helps you search ,match,and manipulate string using pattern
```

```python
# Example :
import re
text ="Email me at vikas jethava VikasJethava@gmail.com"
match = re.search(r"\b[\w.-]+@[\w.-]+\.\w+\b",text)  
# r"...” — raw string literal so backslashes are treated literally.

# \b — word boundary (start/end of word).
# [\w.-]+ — one or more word characters (A-Za-z0-9_), dots . or hyphens - (local part).
# @ — literal at-sign.
# [\w.-]+ — domain name chunk (subdomain/domain).
# \. — literal dot before TLD.
# \w+ — top-level domain (like com, org; letters/digits/underscore).
# \b — word boundary after the TLD.

if match :
    print("Found",match.group()) # match.group() returns the actual substring matched by the pattern (the email).
```
