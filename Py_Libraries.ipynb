3. Matplotlib :Data Visulization

```python
# Matplotlib helps you create charts and plots.
```

```python
# Example:
import matplotlib.pyplot as plt
x=["vikas","sanket","vishal"]
y=[8,10,17]
plt.plot(x,y)
plt.title("Simple Line Plot")
plt.show()
```

4. Seaborn : Statistical Plots

```python
# Seaborn builds on matplotlib with beatiful.high-level charts
```

```python
import seaborn as sns
import pandas as pd
tips = sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=tips)
plt.show()
```

Challenge : Create a heatmap of correlation 

```python
corr =tips.corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```

6.Beautiful Soup :Web Scriping

```python
# Beautiful soup helps you extract data from html pages.
```

```python
from bs4 import BeautifulSoup
import requests

url="https://www.youtube.com/shorts/AdrHkAmliu0"
html=requests.get(url).text
soup=BeautifulSoup(html,"html.parser")
print(soup.title.text)
```
