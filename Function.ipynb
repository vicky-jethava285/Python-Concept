4. Varibale-Length Argument

```python
# *Use *Args and *kwargs for flexibility
```

```python
def add_all(*args):
    return sum(args)
print(add_all(10,24,33,41))
```

```python
def show_info(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

show_info(name="vikas",age=21)
```

6. Return Multiple Values

```python
# Function can return tuples
```

```python
def status(number):
    return min(number),max(number)
low,high=status([3,7,1,9])
print(f"low number is {low} and high number is {high}")
```

```python
# 1. check odd even number
# n=int(input("enter a number"))
def is_even(n):
    return n % 2 ==0
print(is_even(5))
```

```python
# 2. Create a function that return the factorial of a number

def factorial(n):
    result = 1
    for i in range(2,n+1):
        result *=i
    return result
print(factorial(5)) # 1*2*3*4*5
```

```python
# 3. write a function that reverse a string python

def reverse_String(s):
    return s[::-1]
name=input("Enter your name :")
print(reverse_String(name))
```

```python
# 4. write a function that counts vowels in a string
def count_vowels(s):
    return sum(1 for char in s.lower()if char in "aeiou")
print(count_vowels("ioraeu"))
```

```python
# 5. write a function that retrun the nth fibonacci number

def fibo(n):
    a,b=0,1
    for _ in range(n):
        a,b = b,a+b
    return a
print(fibo(4))
```
